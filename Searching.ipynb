```python
from time import time
import random
import numpy as np
import matplotlib.pyplot as plt

class BinarySearch:
    """Class that represents a Binary Search implementation."""
    def __init__(self,items,target):
        self.items=items
        self.target=target
    def _search(self):
        # your code here
        left, right = 0, len(self.items) - 1

        while left <= right:
            mid = (left + right) // 2
            if self.items[mid] == self.target:
                return mid
            elif self.items[mid] < self.target:
                left = mid + 1
            else:
                right = mid - 1

        return -1  # Return -1 if the target is not found

    def _time(self):
        # your code here
        start = time()
        index = self._search()
        end = time()
        self.time = (end - start)
        return self.time
```

```python
class LinearSearch:
    def __init__(self,items,target):
        self.items=items
        self.target=target
    def _search(self):
        for i in range(len(self.items)):
          if self.items[i] == self.target:
             return i
        return -1
    def _time(self):
        start=time()
        self._search()
        end=time()
        self.time = end-start
        return self.time
```

```python
LinearSearch([1,3,14],14)._search()
```

```python
LinearSearch([1,3,14,100],104)._search()
```

```python
BinarySearch([1,4,31,111,555],555)._time()
```

```python
random.seed(42)
test_cases=[]
for i in range(10,201):
    test_cases.append([random.randint(1,1000) for _ in range(i)])
```

```python
sorted_test_cases=[]
for i in test_cases:
  sorted_test_cases.append(sorted(i))
```

```python
target=[]
for i in test_cases:
  target.append(np.random.choice(i,1))
```

```python
linear_search_time=[]
for i,j in zip(test_cases,target):
  linear_search_time.append(LinearSearch(i,j)._time())
```

```python
binary_search_time=[]
for i,j in zip(sorted_test_cases,target):
  binary_search_time.append(BinarySearch(i,j)._time())
```

```python
plt.plot(range(10,201),binary_search_time,label='Binary Search')
plt.plot(range(10,201),linear_search_time,label='Linear Search')
plt.xlabel('List of Varying Size from 10 to 200')
plt.legend()
```
